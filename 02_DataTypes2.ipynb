### **Exercise 1: Changing Data Type**

**Task**:
Write a function `show_details()` that:

* Prompts the user for:

  * **name** (string)
  * **age** (integer)
  * **height in metres** (float)

* Prints:
  `Hello name you are age years old and height metres tall`

> Ensure correct data type conversion using `int()` and `float()`.

**Test Input**:

```
Jean  
52  
2.3
```

**Expected Output**:

```
Hello Jean you are 52 years old and 2.3 metres tall
```

---

````python
def show_details():
    name:input('Enter your name')
    age:int(input('Enter your age'))
    height:float(input('Enter your height in meters'))
    print('Hello ' + name + ', you are ' + age + ' years old.' and ' your height is ' + height + ' meters tall.')

show_details()
````

### **Exercise 3: Converting Float to Int**

**Task**:
Write a function `count_pound_coins()` that:

* Asks the user to enter an amount in **pounds and pence**
* Stores it in a float variable: `pounds_pence`
* Converts it to an integer and stores in `pound_coins`
* Prints:
  `Number of pound coins pound_coins`

> Use `int()` to convert the float to an integer, discarding the decimal part.

**Test Input**:

```
3.48
```

**Expected Output**:

```
Number of pound coins 3
```

---

````python
def count_pound_coins():
    pounds_pence=float(input('Enter the amount in pounds and pence : '))
    pound_coins=int(pounds_pence)
    print("Number of pound coins, pound_coins:", pound_coins)

    count_pound_coins()
````

### **Exercise 5: Area of a Circle**

**Task**:
Write a function `calculate_area()` that:

* Asks the user to enter the **radius** of a circle (can be `int` or `float`)
* Stores the value in a variable `radius`
* Defines a variable `pi = 3.14159`
* Calculates area:
  `area = pi * radius * radius`
* Prints the area **rounded to 2 decimal places** using:
  `print(round(area, 2))`

> Use `float()` to handle both integer and decimal inputs for radius.

**Test Input**:

```
5
```

**Expected Output**:

```
78.54
```

---

````python
def calculate_area():
    radius = float(input("Enter the radius of the circle: "))
    pi = 3.14159
    area = pi * radius * radius
    print(round(area, 2))

calculate_area()
````
