### ***Exercise 9 - Modulus***

**Task:**  
Write a function called `get_pm_minutes()` which will:

- Ask the user to enter an afternoon time in **24-hour clock format** (e.g. `1350`).
- Use **floor division** (`//`) to extract the hour (i.e., divide by 100).
- Use the **modulus operator** (`%`) to extract the minutes.
- Convert the hour to **12-hour clock** by subtracting 12 and store the result back in `hour`.
- Print a message in the format:  
  `"It is {minutes} minutes past {hour} PM"`

**Test Input:**
```
1350
```
**Expected Output:**
```
It is 50 minutes past 1 PM

```
> ðŸ’¡ *Note:* This code is designed for **afternoon (PM)** times. For morning times (AM), additional conditional logic will be required, which you will explore later.
```

````python
def get_pm_minutes():
    time_24 = int(input("Enter the time in 24-hour format (e.g., 1350): "))
    hour = time_24 // 100
    minutes = time_24 % 100
    hour_12 = hour - 12  # converting to 12-hour format for PM
    print(f"It is {minutes} minutes past {hour_12} PM")

# Test the function
get_pm_minutes()
````

### ***Exercise 10 - How Many Tins of Beans?***

**Task:**  
Write a function called `calculate_tins()` that will calculate how many **tins of beans** can fit in a cardboard box.

**Details:**

- Each bean tin is **11 cm tall** and **8 cm in diameter**.
- All tins will be **standing upright** and can be **stacked in layers**.
- Use the **height, width, and depth** of the box supplied in the functionâ€™s parameters.
- Calculate:
  - The **number of layers** (based on box height and tin height)
  - The **number of tins per row and column** (based on box width and depth vs tin diameter)
  - Multiply all three to get the **total number of tins** that can fit.

**Instructions:**

- Ensure all answers are **whole numbers**.
- Print how many tins will fit in the box.

**Hint:**  
Use floor division to ensure youâ€™re only counting full tins that fit.

**Expected Output:**
```

24 tins will fit in this box

```

````python
def calculate_tins(height, width, depth):
    tin_height = 11
    tin_diameter = 8

    layers = height // tin_height
    tins_along_width = width // tin_diameter
    tins_along_depth = depth // tin_diameter

    number_of_tins = layers * tins_along_width * tins_along_depth
    print(f"{number_of_tins} tins will fit in this box")

# Test the function
calculate_tins(24, 50, 18)
````
