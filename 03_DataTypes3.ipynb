### **Exercise 6: Capitalizing Names**

**Task**
Write a function named `capitalise_names()` that:

* Prompts the user to input a **name** and a **surname**.
* Converts both to lowercase, then capitalizes the **first letter** of each using the `.capitalize()` method.
* Prints the formatted name and surname with only the first letters in uppercase.

---

**Test Input**:

```
benJamIN  
smith
```

**Expected Output**:

```
Benjamin Smith
```

> ðŸ’¡ Use `.capitalize()` to ensure only the first letter is uppercase and the rest are lowercase.

---

````python
def capitalise_names():
    name = input('Enter your name: ')
    surname = input('Enter your surname: ')
    formatted_name = name.capitalize()
    formatted_surname = surname.capitalize()
    print(f'Your name is: {formatted_name} {formatted_surname}')

# Test the function
capitalise_names()
````

**Task**
Write a function named `show_postcode_letters()` that:

* Prompts the user to enter a **postcode**.

* Converts `postcode_area` to **uppercase** and stores it in a variable called `capitalised_area`.

---

**Test Input**:


**Expected Output**:

### **Exercise 9: String Repetition**

**Task**
Write a function named `repeat_two_letters()` that will:

* Prompt the user to enter a **word**.
* Assign the **last two letters** of the word to a variable named `last_two`.
* Create a string of **5 copies** of `last_two` (using `last_two * 5`) and assign it to a variable called `five_copies`.
* Print the value of `five_copies`.

**Test Input**:

```
Data
```

**Expected Output**:

```
tatatatata
```


````python
def repeat_two_letters():
    word:str(input('Enter a word:'))
    last_two:str(word[-2:])
    five_copies:str(last_two*5)
    print(f'The last two letters of the word are: {last_two} and they are repeated 5 times: {five_copies}')
````

### **Exercise 10: Converting String to Uppercase for Substring**

**Task**
Write a function named `convert_to_lower_threeLetters()` that will:

* Assign the value `"january"` to a variable called `month`.
* Assign the **first three characters** of `month` to a variable called `month_short`.
* Convert `month_short` to **uppercase** and store it back into `month_short`.
* Assign the **remaining characters** of `month` to a variable called `month_rest`.
* Join `month_short` and `month_rest` and store the result back in `month`.
* Print the final `month`.

**Expected Output**:

```
JANuary
```

````python
def convert_to_lower_threeLetters():
    month:'January'
    month_short:month[0:3]
    month_short:month_short.upper()
    month_rest:month[3:]
    month:str(month_short+month_rest)
    print(f'The month is: {month}')
````
